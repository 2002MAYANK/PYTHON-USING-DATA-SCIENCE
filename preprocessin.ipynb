```python
def clean_price_column(price):
    if isinstance(price, str):
        price = price.replace(',',"")
        price = float(price)
        return price
    return price
```

```python
def clean_price_column(rating):
    if isinstance(rating, str):
        if len(rating) <=3:
            rating = float(rating)
            return rating
        else:
            return np.nan
        return rating
```

```python
# test
clean_price_column('nop review')
```

```python
# testing the fuction on the column
df['rating'].apply(clean_price_column)
```

```python
# implementing the function on the column
df['rating'] = df['rating'].apply(clean_price_column)
df
```

```python
def clean_price_column(price):
    if isinstance(price, str):
        price = price.replace(',',"").replace('₹','')
        price = float(price)
        return price
    return price
```

```python
df['clean_oprice'] = df['oprice'].apply(clean_price_column)
df
```
